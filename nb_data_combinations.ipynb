## Combinations in Python

```python
from numpy import random as r
import datetime
import pandas
from itertools import combinations
```

### Create test data

```python
cats = ["blue","green","orange","purple","black",
        "grey","white","teal","aqua","red","yellow",
        "bone","rainbow",'chrome','silver']


data = pandas.DataFrame(columns = ['unique', 'cats'])

for n in range(2200):
    
    data.loc[n, 'unique'] = 'uniq_' + str(n)
    
    string = ''
    
    for x in range(r.randint(1, len(cats))):
        r2 = r.randint(0, len(cats))
        string = string + cats[r2] + ','
        
    data.loc[n, 'cats'] = string
```

### Get unique terms

```python
uniq_cats = sorted(list(data['cats'].unique()))
```

```python
uniq_cats
```

```python
uniq_terms = []
delimeter = ','

for item in uniq_cats:
    
    split = item.split(delimeter)
    for term in split:
        if len(term) > 0 and term not in uniq_terms:
            uniq_terms.append(term)
```

```python
uniq_terms = sorted(uniq_terms)
uniq_terms
```

### Combinations from `itertools`

```python
combos = list(combinations(uniq_terms, 2))
doub_list = sorted([(y, x) for x, y in combos] + combos)
```

```python
for x,y in enumerate(combos):
    first = y[0]
    second = y[1] 
    subset = data[data['cats'].str.contains(first) & data['cats'].str.contains(second)]
    #print(x,y,subset.shape)
```
