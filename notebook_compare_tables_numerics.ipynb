```python
relevant_columns = list(old_table.columns)
```

```python
dictionary = {}

wrong = 0
total = old_table.shape[0] * len(relevant_columns)

for label in relevant_columns:
    dictionary[label] = []

for rowlabel in range(old_table.shape[0]):

    for columnlabel in relevant_columns:
        code = old_table.loc[rowlabel,"code"]
        old_value = round(old_table.loc[rowlabel, columnlabel],1)
        new_value = round(new_table.loc[rowlabel, columnlabel],1)

        if old_value == new_value:
            match_flag = True
        else:
            match_flag = False
            wrong = wrong + 1
            print(match_flag,code,columnlabel,old_value,new_value,sep = " : ")

        dictionary[columnlabel].append(match_flag)
```

```python
print("{:.2%}".format(wrong/total)," of values in new table did not match previous data set.")
```

```python
data = pandas.DataFrame(dictionary)
```
